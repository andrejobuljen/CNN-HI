## Which params

Let's consider cosmological:
- $\Omega_m = [0.1,0.5]$
- $\sigma_8 = [0.6,1]$

and nuisance parameters:

- $b_1 = [0.5,2]$
- $b_2 = [-1,1]$
- $b_{\mathcal{G}_2} = [-1,1]$
- $P_\epsilon = [80, 1000]$

```python
params = ['om', 's8', 'b1', 'b2', 'bg2', 'Pn']
len(params)
```

```python
om_lims = np.array([0.1, 0.5])
s8_lims = np.array([0.6, 1.0])
b1_lims = np.array([0.5, 2.0])
b2_lims = np.array([-1.0, 1.0])
bG2_lims = np.array([-1.0, 1.0])
Psn_lims = np.array([80, 1000])
print (np.diff(om_lims), np.diff(s8_lims), np.diff(b1_lims), np.diff(b2_lims), np.diff(bG2_lims), np.diff(Psn_lims))
```

```python
sampler = qmc.Sobol(d=len(params), scramble=True, seed=500)
sample = sampler.random_base2(m=10)

print (sample[:,-1].min(), sample[:,-1].max())
```

All the params are between 0-1

```python
sample[:,0] *= np.diff(om_lims)
sample[:,0] += om_lims[0]

sample[:,1] *= np.diff(s8_lims)
sample[:,1] += s8_lims[0]

sample[:,2] *= np.diff(b1_lims)
sample[:,2] += b1_lims[0]

sample[:,3] *= np.diff(b2_lims)
sample[:,3] += b2_lims[0]

sample[:,4] *= np.diff(bG2_lims)
sample[:,4] += bG2_lims[0]

sample[:,5] *= np.diff(Psn_lims)
sample[:,5] += Psn_lims[0]
```
